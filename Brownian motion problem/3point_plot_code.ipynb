```python
import random
import numpy as np
import matplotlib.pyplot as plt
radius = 55
theta = np.linspace( 0 , 2 * np.pi)
a = radius * np.cos( theta )
b = radius * np.sin( theta )
figure, axes = plt.subplots( 1 )

p = [-70, 70, 70, -70, -70]
q = [-70, -70, 70, 70, -70]

z1 = random.uniform(0,np.pi)
x1 = radius*np.cos(z1)
y1 = radius*np.sin(z1)

x1_coords = [x1]
y1_coords = [y1]

for i in range(1000):
    if random.choice([True, False]):
        x1 += random.choice([-1, 1])
        if x1>70:
           x1=-(140-x1)
        elif x1<-70:
           x1=140-abs(x1)
        else:
            x1=x1
    else:
        y1 += random.choice([-1, 1])
        if y1>70:
           y1=-(140-y1)
        elif y1<-70:
           y1=140-abs(y1)
        else:
            y1=y1
    x1_coords.append(x1)
    y1_coords.append(y1)
    
z2 = random.uniform(0,np.pi)
x2 = radius*np.cos(z2)
y2 = radius*np.sin(z2)

x2_coords = [x2]
y2_coords = [y2]

for i in range(1000):
    if random.choice([True, False]):
        x2 += random.choice([-1, 1])
        if x2>70:
           x2=-(140-x2)
        elif x2<-70:
           x2=140-abs(x2)
        else:
            x2=x2
    else:
        y2 += random.choice([-1, 1])
        if y2>70:
           y2=-(140-y2)
        elif y2<-70:
           y2=140-abs(y2)
        else:
            y2=y2
    x2_coords.append(x2)
    y2_coords.append(y2)
    
z3 = random.uniform(0,np.pi)
x3 = radius*np.cos(z3)
y3 = radius*np.sin(z3)

x3_coords = [x3]
y3_coords = [y3]

for i in range(1000):
    if random.choice([True, False]):
        x3 += random.choice([-1, 1])
        if x3>70:
           x3=-(140-x3)
        elif x3<-70:
           x3=140-abs(x3)
        else:
            x3=x3
    else:
        y3 += random.choice([-1, 1])
        if y3>70:
           y3=-(140-y3)
        elif y3<-70:
           y3=140-abs(y3)
        else:
            y3=y3
    x3_coords.append(x3)
    y3_coords.append(y3)
    
    
plt.subplot()
axes.plot( a, b )
plt.plot(x1_coords, y1_coords,color="red")
plt.plot(x2_coords, y2_coords,color="orange")
plt.plot(x3_coords, y3_coords,color="green")
axes.set_aspect( 1 )
plt.plot(p, q, color="black")
plt.axis('off')
plt.savefig("plot_3_particles.jpg")

plt.show()



import csv

with open('point_coord1.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(zip(x1_coords, y1_coords))
with open('point_coord2.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(zip(x2_coords, y2_coords))
with open('point_coord3.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(zip(x3_coords, y3_coords))    
```
